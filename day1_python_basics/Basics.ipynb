```python
# Cell 41: making a mini phone book dictionary
phone_book = {"mom": 1234, "dad": 5678}
print(phone_book)
```

```python
# Cell 43: adding a new person to dictionary
phone_book["bro"] = 9999
print(phone_book)
```

```python
# Cell 46: simple if statement trial
marks = 75
if marks >= 50:
    print("Passed!")
```

```python
# Cell 47: testing else option
temperature = 15
if temperature > 30:
    print("Hot")
else:
    print("Not hot")
```

```python
# Cell 48: testing elif mid conditions
speed = 45
if speed > 60:
    print("Too fast")
elif speed < 30:
    print("Too slow")
else:
    print("Perfect speed")
```

```python
# Cell 49: combining 'not' with if statement
is_weekend = True
if not is_weekend:
    print("Go to school")
else:
    print("Sleep in")
```

```python
# Cell 50: using remainder math to check odd or even numbers
num = 15
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Cell 51: simple for loop printing items
for num in [1, 2, 3]:
    print("Loop count: " + str(num))
```

```python
# Cell 52: adding numbers up inside a loop
total_sum = 0
for n in [5, 10, 15]:
    total_sum = total_sum + n
print(total_sum)
```

```python
# Cell 53: while loop running until count hits 3
count = 0
while count < 3:
    print("Count is " + str(count))
    count = count + 1
```

```python
# Cell 54: looping through characters of a string word
word = "abc"
for letter in word:
    print(letter)
```

```python
# Cell 55: making a basic function with no input
def say_hi():
    print("Hi there!")
say_hi()
```

```python
# Cell 56: function that multiplies input by 2
def double_me(val):
    print(val * 2)
double_me(10)
```

```python
# Cell 57: function using return keyword to give back answer
def add_nums(a, b):
    return a + b
result = add_nums(4, 5)
print(result)
```

```python
# Cell 58: putting an if statement inside a function
def check_voter(age):
    if age >= 18:
        return True
    else:
        return False
print(check_voter(16))
```

```python
# Cell 59: trying to catch a crash using try except
try:
    print(unmade_variable)
except:
    print("Oops that variable doesn't exist yet!")
```

```python
# Cell 60: catching an index crash block
try:
    bad_list = [1, 2]
    print(bad_list[5])
except:
    print("Index out of bounds caught!")
```
